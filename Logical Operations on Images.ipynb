```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img1 = cv2.imread('/content/Letter - 1.png')
img2 = cv2.imread('/content/Letter - 2.png')

# Plot original images
originals = {
    'Original Image 1': img1,
    'Original Image 2': img2
}

plt.figure(figsize=(10, 4))
for i, (k, v) in enumerate(originals.items()):
    plt.subplot(1, 2, i + 1)
    plt.imshow(v[:, :, ::-1])
    plt.title(k)
    plt.xticks([]); plt.yticks([])
plt.tight_layout()
plt.show()

# AND
img_and = cv2.bitwise_and(img1, img2)

# OR
img_or = cv2.bitwise_or(img1, img2)

# NOT
img_not = cv2.bitwise_not(img1)

# Plot images

images = [img_and, img_or, img_not]
titles = ['AND Image', 'OR Image', 'NOT Image (1)']

plt.figure(figsize=(18, 4))

for i in range(len(images)):
    plt.subplot(1, 5, i + 1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')
plt.tight_layout()
plt.show()
```
